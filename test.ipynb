```python
class Coin:
    # The constructor method in Python is __init__
    def __init__(self, name, symbol, value):
        self.name = name          # Instance attribute
        self.symbol = symbol      # Instance attribute
        self.value = value        # Instance attribute
        print(f"Coin '{self.name}' created successfully.")

    # A helper method to display details
    def display_info(self):
        return f"{self.name} ({self.symbol}) - Value: ${self.value}"

# Instantiating an object of the Coin class
btc = Coin("Bitcoin", "BTC", 67000)
print(btc.display_info())
```
